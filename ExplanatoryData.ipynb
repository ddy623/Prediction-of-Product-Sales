```python
#import pandas 
import pandas as pd
#save file path
fname = "https://docs.google.com/spreadsheets/d/e/2PACX-1vS6Sn9LaMSc_E1EHQpuRK6BTpKp6h27obTP_dTpAVu_xtoqsge30jBGh9vYlO4DYe-utRKMgMqYChU_/pub?output=csv"
```

```python
# load in the data as a pandas dataframe
df = pd.read_csv(fname)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



```

```python
fig,ax = plt.subplots(figsize=(20,16))
sns.histplot(data=df, x='Item_Type', ax = ax);
plt.title('Number of Items')
```

```python
# Find the mean and round to 2 decimal places for readability
mean = df['Item_MRP'].mean().round(2)
mean
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(data=df, x = 'Item_MRP', ax = ax)
ax.axvline(mean);
plt.title("Average of MRP")
```

```python
# Create 2 side by side axes with a shared y axis
fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(30,25))
# Plot the barplot on the left
sns.barplot(data=df, x='Item_Type', y='Item_MRP',ax=axes[0]);
# Plot the stripplot on the right
sns.stripplot(data=df, x='Item_Type', y='Item_MRP',ax=axes[1],
                   hue='Item_Type', edgecolor='white', linewidth=1);
plt.title('Grocery Prices per Item Type')

```
