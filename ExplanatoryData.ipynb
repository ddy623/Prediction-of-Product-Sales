Clean the Data

```python
##Check Item_Fat_Content
df['Item_Fat_Content'].value_counts()
```

```python
##replace inconsistent categories
fat_content_map = {'LF': 'Low Fat', 'reg': 'Regular', 'low fat': 'Low Fat'}
df['Item_Fat_Content'] = df ['Item_Fat_Content'].replace(fat_content_map)

##Verify
df['Item_Fat_Content'].value_counts()
```

Explanatory Data Analysis

What is the relationship between Item Fat Content and Item Visibility?

```python
## Define MatplotLib fig and ax objects using plt.subplots()
fig, ax = plt.subplots()

## Define a MatplotLib ax object using sns.scatterplot()
## Using y= to plot the y axis (vertical) values
## Using x= to plot the x axis (horizontal) values
## Using hue= to differential an additional feature
ax = sns.scatterplot(data=df, 
                     x='Item_Fat_Content', 
                     y='Item_Visibility', 
                     hue='Item_Type')

## Set the Title
ax.set_title('Item_Fat_Content vs Item_Visibility', fontsize=12, fontweight='bold')
## Set the X axis label
ax.set_xlabel('Item_Fat_Content', fontsize=10, fontweight='bold')
## Set the Y axis label
ax.set_ylabel('Item_Visibility', fontsize=10, fontweight='bold')

## Set the legend location to [1,1]
ax.legend(bbox_to_anchor=[1,1]);
```

```python
## Define MatplotLib fig and ax objects using plt.subplots()
fig, ax = plt.subplots()

## Define a MatplotLib ax object using sns.barplot()
sns.barplot(data = df, x = 'Item_Fat_Content', y = 'Item_Visibility');

## Set the Title
ax.set_title('Item Fat Content vs Item Visibility', fontsize=12, fontweight='bold')
## Set the X axis label
ax.set_xlabel('Item_Fat_Content', fontsize=10, fontweight='bold')
## Set the Y axis label
ax.set_ylabel('Item_Visibility', fontsize=10, fontweight='bold');
```

What is the relationship between Item Weight and Item MRP?

```python
## Define MatplotLib fig and ax objects using plt.subplots()
fig, axes = plt.subplots(nrows = 2, figsize = (5, 8))

## Define a MatplotLib ax object using sns.barplot() for ax=axes[0]
sns.barplot(data = df, 
            x='Item_Weight', 
            y='Item_MRP', 
            ax = axes[0])
## Set the Title
axes[0].set_title('Item Weight vs Item MRP', fontsize=10, fontweight='bold')
## Set the X axis label
axes[0].set_xlabel('Item_Weight', fontsize=10, fontweight='bold')
## Set the Y axis label
axes[0].set_ylabel('Item_MRP', fontsize=10, fontweight='bold')

## Define a MatplotLib ax object using sns.barplot() for ax=axes[1]
sns.barplot(data=df, 
            x='Item_Weight',
            y='Item_MRP', 
            ax = axes[1])
## Set the Title
axes[1].set_title('Item Weight vs. Item_MRP', fontsize = 10, fontweight = 'bold')
## Set the X axis label
axes[1].set_xlabel('Item_Weight', fontsize=10, fontweight='bold')
## Set the Y axis label
axes[1].set_ylabel('Item_Outlet_Sales', fontsize=10, fontweight='bold')

## Set the Suptitle
fig.suptitle('Impact on Item_Outlet_Sales vs MRP', fontsize = 12, fontweight = 'bold', y = .99)

## Set tight layout
fig.tight_layout();
```
